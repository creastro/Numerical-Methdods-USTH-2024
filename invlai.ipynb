## Ex 6
Create a new figure (Figure 2) and plot on the same graph the six <i>NDVI</i> curves obtained in Question 5 as a function of <i>LAI</i> . The x-axis should be between 0 and 10 and the y-axis between 0 and 1. Add appropriate labels to the x-axis and a legend.

```python
lai = np.arange(0,10.1,0.1)
K = np.arange(0.5, 1.6, 0.2)
ndvi_inf = 0.8
ndvi_soil = 0.1

for k in K:
    plt.plot(lai,modlai(lai,ndvi_inf,ndvi_soil,k), label = f'K = {np.round(k,2)}')
    
plt.legend()
plt.xlabel('Leaf Area Index ($m^2/m^2$)')
plt.ylabel('NDVI')
plt.yticks(np.arange(0,1.1,0.1))
plt.title('Evolution of NVDI as a function of LAI')
```

## Ex 8
Let’s consider soybean crop. In the main program, `invlai`, determine the set of optimal parameters <b>res</b> that minimizes the merit function using the `scipy.optimize.fmin` routine. Assume the initial guess <b>Pi=[1 1 1]</b>. Then define the vector <b>lai_mod</b> (<i>LAI</i>=0...7 with $\Delta$<i>LAI</i>=0.1) and calculate <b>ndvi_mod</b> the <i>NDVI</i> for the soybean crop, using the values of the retrieved parameters <b>res</b>.

```python
from scipy.optimize import fmin
```

```python
res2 = fmin(chi2lai, [1, 1, 1], args=(laindvi2['lai2'],laindvi2['ndvi2']))
```

```python
res2
```

## Ex 9
Create a new figure (Figure 3) and plot on the same graph <b>ndvi2</b> as a function of <b>lai2</b> (black stars), and <b>ndvi_mod</b> as a function of <b>lai_mod</b> (red line). Add the appropriate labels to the x-axis and the y-axis, « Measure » and « Model » in the legend. The retrieved values of the three parameters will be automatically displayed in the graph, for example in the title. The x-axis must lie between 0 and 7 and the y-axis between 0 and 1.

```python
lai_mod_2 = np.arange(0,7.1,0.1)
ndvi_mod_2 = modlai(lai_mod_2, res2[0], res2[1], res2[2])
```

```python
plt.scatter(laindvi2['lai2'],laindvi2['ndvi2'], marker = '*', color = 'black', s = 100, label = 'Measure')
plt.plot(lai_mod_2, ndvi_mod_2, 'r', label = 'Model')

plt.xlabel('Leaf Area Index')
plt.ylabel('NDVI')
plt.xticks(np.arange(0,8))
plt.yticks(np.arange(0,1.1,0.1))

plt.legend()
```
